```python
import pandas as pd
import numpy as np
import os
```

```python
data=pd.read_csv('tmdb-movies.csv')
data.columns
```

```python
movie=data[['original_title','cast','director','keywords','genres']]
```

```python
movie.head()
```

```python
# Remove null values 
movie=movie.fillna('')
```

```python
#Remove | from columns cast, director, keywords and genres

columns=['cast','director','keywords','genres']

for i in columns:
    movie[i]=movie[i].str.replace('|',' ')
```

```python
movie
```

```python
#lower case all the features
features=['original_title','cast','director','keywords','genres']
for i in features:
    movie[i]=movie[i].str.lower()
```

```python
movie.head()
```

```python
movie['comb']=movie['keywords']+' '+movie['genres']+' '+movie['director']+' '+movie['cast']
```

```python
movie.rename(columns={"original_title": "movie_title"},inplace=True)
```

```python
movie.to_csv('main_data1.csv',index=False)
```
